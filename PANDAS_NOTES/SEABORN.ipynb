```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.get_dataset_names()
```

### BARPLOT

```python
plt.figure(figsize=(6,3))
sns.barplot(x="species", y="sepal_length", data=iris, palette="Spectral")
plt.plot()
```

```python
mtcars = pd.read_csv(".//DATA//mtcars.csv")
mtcars.head()
```

```python
groupby_cyl = mtcars.groupby("cyl")
groupby_cyl["mpg"].mean()
```

```python
plt.figure(figsize=(6,3))
sns.barplot(x="cyl", y="mpg", hue="am", data=mtcars, palette="Set2")
plt.plot()
```

### COUNTPLOT

```python
sns.set_style("darkgrid")
sns.set(rc={'figure.figsize':(7,4)})
sns.countplot(x="cyl", hue="am", data=mtcars, palette="coolwarm")
plt.plot()
```

```python
sns.countplot(x="cyl", hue="gear", data=mtcars, palette="Set2")
plt.plot()
```

### BOXPLOT

```python
sns.set_style("ticks")
sns.boxplot(x="cyl", y="mpg", hue="am", data=mtcars, palette="Set2")
plt.plot()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips, palette="Blues")
plt.plot()
```

### VIOLINPLOT

```python
sns.violinplot(x="cyl", y="mpg", hue="am", data=mtcars, palette="Set2")
plt.plot()
```

```python
sns.set_style("ticks")
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, palette="Blues")
plt.title("VIOLIN PLOT")
plt.xlabel('DAY')
plt.ylabel('TOTAL_BILL')
```

### CATPLOT

```python
sns.set_style("ticks")
df = sns.load_dataset("titanic")
df.head(10)
```

```python
sns.catplot(data=df, x="age", y="class")
```

```python
sns.catplot(data=df, x="age", y="class", kind="box")
```

```python
sns.catplot(
    data=df, x="age", y="class", hue="sex",
    kind="violin", bw=.25, cut=0, split=True,
)
```

```python
sns.catplot(
    data=df, x="class", y="survived", col="sex",
    kind="bar", height=4, aspect=.6,
)
```

### LINEPLOT

```python
flights = sns.load_dataset("flights")
flights.head(), flights["month"].unique()
```

```python
sns.set_style("darkgrid")
may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x="year", y="passengers")
```

```python
data_flights = flights.loc[flights["month"].isin(["Oct", "Jan", "Feb", "Mar"])]
sns.lineplot(data=data_flights, 
             x="year", y="passengers", hue="month", style="month", size="month")
```

```python
fmri = sns.load_dataset("fmri")
print(fmri.head())
sns.lineplot(
    data=fmri, x="timepoint", y="signal", hue="event", err_style="bars", errorbar=("se", 2),
)
```
