```python
import pandas as pd
import seaborn as sns
tendulkar_data = pd.read_csv(".//DATA//tendulkar_ODI.csv")
tendulkar_data
```

```python
def remove_star(x):
    if "*" in x:
        x=x[:-1]
    return x
pd.set_option('display.max_rows', 200)
runs = tendulkar_data["Runs"].apply(remove_star).value_counts()
```

```python
type(runs), runs.index, runs.values
```

```python
sns.barplot(x=runs.index[:10], y=runs.values[:10])
```

# Segmented Analysis

```python
EDA_data = pd.read_csv(".//DATA//EDA_nas.csv")
print(EDA_data.columns)
EDA_data = EDA_data.fillna(0)
EDA_data
```

```python
play_games_group = EDA_data.groupby(["Play.games"])
play_games_group["Reading.."].mean()
```

```python
watch_tv_group = EDA_data.groupby(["Watch.TV"])
watch_tv_group["Science.."].mean(), watch_tv_group["Science.."].max()
```

```python
watch_tv_group = EDA_data.groupby(["Watch.TV", "Gender"])
watch_tv_group["Science.."].mean(), watch_tv_group["Maths.."].mean()
```

```python
EDA_GOLD_SILVER_PRICES = pd.read_csv(".//DATA//EDA_Gold_Silver_prices.csv")
print(EDA_GOLD_SILVER_PRICES.head())
print(EDA_GOLD_SILVER_PRICES.columns)
print("Correlation btw gold & silver prices is : ",EDA_GOLD_SILVER_PRICES["SilverPrice"].corr(EDA_GOLD_SILVER_PRICES["GoldPrice"]))
```
